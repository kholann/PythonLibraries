```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
plt.style.use('fivethirtyeight')
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
creditcard = pd.read_csv('creditcard.csv')
```

```python
class_list = creditcard['Class'].value_counts()
print(class_list)
```

```python
class_list.plot(kind='barh')
plt.show()
```

```python
class_list.plot(kind='barh', logx=True)
plt.show()
```

```python
class_regular = creditcard.loc[creditcard['Class'] == 0, ['V1']]
class_fraud = creditcard.loc[creditcard['Class'] == 1, ['V1']]
```

```python
plt.hist(class_regular['V1'], bins=20, density=True, alpha=0.5, label='Class 0', color='grey')
plt.hist(class_fraud['V1'], bins=20, density=True, alpha=0.5, label='Class 1', color='red')
plt.xlabel('V1')
plt.legend()
plt.show()
```
